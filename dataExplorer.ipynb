```python
bins_100 = [b for b in range(150,1000,50)]
bins_500 = [b for b in range(1000,10000,500)]
bins_10000 = [b for b in range(10000,50000,10000)]
df['PriceUSD'].value_counts(bins=[*bins_100, *bins_500, *bins_10000], sort=False).to_frame().plot() #.head()
```
